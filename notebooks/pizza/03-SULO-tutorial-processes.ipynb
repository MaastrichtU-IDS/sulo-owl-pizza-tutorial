Making the Dough
- a transformative process
- has flour, salt, yeast, and water as an input
- has a dough as an output
- individual ingredients cease to exist, and the dough comes into existence 

Proofing the Dough
- a non transformative process
- follows the dough preparation
- change in qualities (airness, volume)

Before proofing,	After proofing
Texture	Dense, tight	Soft, aerated
Volume	Small	Expanded (1.5–2×)
Aroma	Neutral flour	Yeasty, slightly acidic
Gluten	Tight and elastic	Relaxed, extensible


Baking the Pizza

The cheese on the pizza participates in a heating process in which its physical state, color, texture, and aroma qualities are transformed through melting, dehydration, and Maillard reactions, producing the characteristic golden, bubbling appearance of baked cheese.

- a transformative process for some of the ingredients (dough converted into pizza crust;
-
- The cheese transitions from a solid to a viscous, molten emulsion. the melting process changes the phase and texture qualities of the cheese without altering its identity as the same spatial object.

Slicing the Pizza


```python
base_iri = "https://w3id.org/ontostart/sulo-ext/"
se = get_ontology(base_iri)

with se:
    ## Object Properties
    class hasDirectParticipant(ObjectProperty):
        """A property that relates a process to its direct participants."""
        pass

    class hasInput(hasDirectParticipant):
        """A property that relates a process to its input entities."""
        pass
    class hasOutput(hasDirectParticipant):
        """A property that relates a process to its output entities."""
        pass
    
    class hasAgent(hasDirectParticipant):
        """A property that relates a process to its agent, typically an entity that intentionally brings about the process."""
        pass

    ## Classes
    class TransformationProcess(sulo.Process):
        """A process that transforms an input entity into an output entity of a different kind or identity."""
        pass
    class DevelopmentalProcess(sulo.Process):
        """A process that brings about changes in the qualities, states, or configurations of an existing object without altering its identity or kind."""
        pass
    class CompositionProcess(sulo.Process):
        """A process that results in the creation of a new entity by combining multiple input entities."""
        pass
    class DecompositionProcess(sulo.Process):
        """A process that results in the breakdown of an entity into its constituent parts."""
        pass
    
```

```python


with pizza:
    class Flour(pizza.FoodMaterial):
        pass

    class Water(pizza.FoodMaterial):
        pass

    class Sugar(pizza.FoodMaterial):
        pass
    
    class BakersYeast(pizza.FoodMaterial):
        pass

    class Salt(pizza.FoodMaterial):
        pass

    class PizzaDough(pizza.FoodMaterial):
        pass

    class MakingThePizzaDough(sulo.Process  ):
        is_a = [sulo.Process
                & pizza.hasInput.exactly(1, Flour)
                & pizza.hasInput.exactly(1, Water)
                & pizza.hasInput.exactly(1, Sugar)
                & pizza.hasInput.exactly(1, BakersYeast)
                & pizza.hasInput.exactly(1, Salt)
                & pizza.hasOutput.exactly(1, pizza.PizzaDough)
                ]
```

DevelopmentalProcess: A process that brings about changes in the qualities, states, or configurations of an existing object without altering its identity or kind.

Formally:
A sulo:Process in which the input and output denote the same individual, whose intrinsic qualities (e.g., shape, color, phase, or maturity) are modified over time.

TransformativeProcess: A process that results in the production of a new object whose identity or kind differs from that of its input.

Formally:
A sulo:Process in which the output is a distinct individual from the input, representing a change in material constitution, structure, or categorical type.
